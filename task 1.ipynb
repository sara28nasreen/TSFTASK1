####  To Predict the percentage of marks of the students based on the number of hours they studied ?

```python
# Importing all the required libraries

import pandas as pd
import numpy as np  
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
import seaborn as sns
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
# Reading the Data file and printing top 5 data

data = pd.read_csv('http://bit.ly/w-data')
print("Data imported!")

data.head()
```

```python
# Check if there any null value in the Dataset
data.isnull == True
```

### There is no null value in the Dataset so, we can now visualize our Data.

```python
#Plotting the Graph of the given data:
# Plotting the Hours vs Percentage graph

data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage', size=20)  
plt.xlabel('No. of hours Studied')  
plt.ylabel('Percentage Scores') 
plt.grid()
plt.show()
```

Plotting it in Linear Regression Graph:

```python
plt.title('Linear Regression Plot', size=20)  
sns.regplot(x= data['Hours'], y= data['Scores'])
plt.ylabel('Marks Percentage')
plt.xlabel('Hours Studied')
plt.grid()
```

### Model Training

```python
# Defining X and y from the Data
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values

# Spliting the Data in two training and test sets
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 0)
```

```python
#Fitting our training set into the model

regr = LinearRegression()
regr.fit(train_X, train_y)
print("Our Model is Trained!")  
```

#### Prediction of the percentage:-

```python
#Predicting the marks - In hours

pred_y = regr.predict(test_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in test_X], 'Predicted Scores': [j for j in pred_y]})
prediction[:6]
```
